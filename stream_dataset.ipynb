```python
import keras_nlp
import numpy as np
from datasets import load_dataset
```

```python
dataset = load_dataset(
    "EleutherAI/the_pile_deduplicated", split="train", streaming=True
)
dataset = iter(dataset)
```

```python
def next_text():
    return next(dataset)["text"]


sample = next_text()
print(sample)
```

```python
tokenizer = keras_nlp.tokenizers.ByteTokenizer(
    lowercase=True,
    sequence_length=None,
    normalization_form="NFKC",
    errors="replace",
    replacement_char=65533,
    dtype="int32",
)


def tokenize(string):
    return list(tokenizer(string).numpy())


tok_sample = tokenize(sample)
print(tok_sample)


def detokenize(data):
    return tokenizer.detokenize(data).numpy().decode("UTF-8")


print(detokenize(tok_sample))
```

```python
target_size = 1024


def data_gen(target_size):
    cache = [0]
    while True:
        while len(cache) < target_size:
            text = next_text()
            cache += tokenize(text) + [0]

        while len(cache) > target_size:
            out, cache = cache[:target_size], cache[target_size:]
            yield np.array(out)


text_gen = data_gen(target_size)

for i in range(10):
    d = next(text_gen)
    print(len(d))


dataset = tf.data.Dataset.from_generator(
    text_gen, output_signature=(tf.TensorSpec(shape=(target_size), dtype=tf.int32),)
)
```
